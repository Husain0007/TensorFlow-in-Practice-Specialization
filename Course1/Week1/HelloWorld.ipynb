```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
# from keras.models import Sequential
# from keras.layers import Dense, Activation
```

### Single Layer Perceptron

```python
model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
# use word 'Dense' to define a layer of connected neurons
# use word 'Sequential' to define successive layers
# 'input_shape' to define shape of what's been input to the neural network
```

```python
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

```python
# Training takes place in the fit command
model.fit(xs, ys, epochs=500)
```

```python
print(model.predict([10.0]))
```
